```python
%matplotlib inline
import pandas as pd
import numpy as np
import os
import stanfordnlp
from tqdm import tqdm
```

```python
path_to_data = '/home/dafne/shared/FilterBubble/topic-modeling/'
fname = os.path.join(path_to_data, 'texts_rens.csv')

data = pd.read_csv(fname)
```

```python
len(data), data.aid.nunique()
```

```python
data[data.aid.isna()]
```

```python
nr_aids = data.groupby('aid').count()['id']
data[data.aid.isin(nr_aids[nr_aids>1].index)]
```

```python
data_deduplicated = data.groupby('aid').first()
```

```python
nlp = stanfordnlp.Pipeline(lang='nl')
```

```python
import warnings
warnings.simplefilter('ignore')

fpath_connl = '/home/dafne/shared/FilterBubble/topic-modeling/parsed_stanfordnlp/'

for aid, row in tqdm(data_deduplicated.iterrows()):
    doc = nlp(row['text'])
    fn_out = os.path.join(fpath_connl, str(aid)+'.connl')
    with open(fn_out, 'w') as f:
        f.write(doc.conll_file.conll_as_string())
```
